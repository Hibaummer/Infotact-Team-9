# Exploratory Data Analysis (EDA)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")

df = pd.read_csv(r"Data/H1_cleaned_feature_engineering.csv")
```

## Univariate Analysis-Numerical

```python
num_cols = df.select_dtypes(include=np.number).columns
for col in num_cols:
    plt.figure()
    sns.histplot(df[col],kde=False)
    plt.title(f"Distribution of {col}")
    plt.show()
```

## Key Insights

##### *1) Rate of cancellations are comparatively low.*
##### *2) Lead time is not normally distributed, most bookings happen with low leadtime.*
##### *3) 2016 has more arrivals.*
##### *4) Arrivals are occurring during 30 to 35 week that is July to August months.*
##### *5) At the end of the month most of the customers are booked Also middle of month.*

## Univariate Analysis-Categorical

```python
#cat_cols = df.select_dtypes(include='object').columns
cat_cols=[]
for col in cat_cols:
    plt.figure(figsize=(8,4))
    df[col].value_counts().head(10).plot(kind='bar')
    plt.title(f'Top categories in {col}')
    plt.show()
```

## Key Insights

##### *1) Highest number of arrivals in August and July, Lowest arrivals recorded in December.*
##### *2) BB is the most preferred meal type, Followed by HB, Very few customers choose FB, SC, or Undefined.*
##### *3) Majority of bookings are from PRT (Portugal), Followed by GBR (United Kingdom) and ESP (Spain), Other countries contribute significantly less.*
##### *4) Online TA has the highest number of bookings, Followed by Offline TA/TO and Direct bookings, Corporate and Groups segments are relatively low.*
##### *5) TA/TO (Travel Agents/Tour Operators) dominate the bookings, Followed by Direct channel, Corporate and Undefined channels have minimal contribution.*
##### *6) Room Type A has the highest number of reservations, Room Types D and E are moderately used, Other room types have very low demand.*
##### *7) Room Type A is most frequently assigned, Room Types D and E follow next, Some differences from reserved rooms indicate room reassignment.*
##### *8) No Deposit is used for almost all bookings, Non-Refund bookings are very few, Refundable option is rarely chosen.* 
##### *9) Transient customers dominate bookings, Transient-Party is the second largest group, Contract and Group customers are minimal.*
##### *10) Most bookings result in Check-Out (successful stay), Cancellations are relatively high, No-shows are very rare.*



## Outlier Detection

```python
for col in num_cols:
    plt.figure()
    sns.boxplot(x=df[col])
    plt.title(f"Outliers in {col}")
    plt.show()
```

##### *ADR have Outliers Lower Quartile -72.5 and Upper Quartile -72.5 ADR rate should be Positive so the Lower is set to 0 and the values above 257.9 are Outliers.*
